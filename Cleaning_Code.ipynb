```python
#creating a copy so that i dont have to reload the csv 
fifa = df.copy()
```

```python
#converting height to cm

def clean_height(x):
    if isinstance(x,str): # only clean if its a string
        if "cm" in x:
            return int(x.replace("cm", ""))
        elif "'" in x:
            feet, inches = x.split("'") # 5
            total_inches = (int(feet)*12)+ int(inches.replace('"', ''))
            return round(total_inches * 2.54 ) # converting to cm
    return x
```

```python
#converting weight to kg  

def clean_weight(x):
    if isinstance(x,str):
        if "kg" in x:
            return int(x.replace("kg", ""))
        elif "lbs" in x:
            lbs = int(x.replace("lbs", ""))
            return round(lbs*0.453)
    return x
```

```python
#conerting money to integer values

def clean_money(x):
    if isinstance(x,str):
        x = x.replace("€", "").replace(" ", "")
        if "M" in x:
            return float(x.replace("M", ""))*1_000_000
        elif "K" in x:
            return float(x.replace("K", ""))*1_000
    return float(x)
```

```python
fifa = fifa.rename(columns={'LongName': 'Full Name', 'OVA': 'Overall'})
```

```python
# dropping duplicates 

fifa = fifa.drop_duplicates()
```

```python
#cleaning all money columns

fifa['Value'] = fifa['Value'].apply(clean_money)
fifa['Wage'] = fifa['Wage'].apply(clean_money)
fifa['Release Clause'] = fifa['Release Clause'].apply(clean_money)
```

```python
fifa[["Value","Wage","Release Clause"]]
```

```python
#cleaning height and weight 

fifa['Height'] = fifa['Height'].apply(clean_height)
fifa['Weight'] = fifa['Weight'].apply(clean_weight)
```

```python
fifa_two =fifa.copy()
```

```python
fifa['Hits'] = fifa_two['Hits'].str.replace('\n', "") 
```

```python
#using lambda because its only one kind of process
# dataset looks like this - 1k 12k 500 3l NaN
# Hits column

fifa['Hits'] = fifa['Hits'].astype(str).apply(lambda x: float(x.replace('K', '')) * 1000 if 'K' in x else x)
fifa['Hits'] = pd.to_numeric(fifa['Hits'], errors='coerce').fillna(0)
```

```python
# Cleaning the messy newlines from the start and end first
# .strip()removes whitespace/newlines from both ends of the string

fifa['Team & Contract'] = fifa['Team & Contract'].astype(str).str.strip()

# 2. Spliting the column into two separate columns
# spliting based on the remaining newline character ('\n') inside the text
# expand=True turns the split results into separate columns immediately
split_columns = fifa['Team & Contract'].str.split('\n', expand=True)

# 3. Assign these new columns back to your dataframe
fifa['Team'] = split_columns[0]
fifa['Contract Duration'] = split_columns[1]

# 4. Drop the original messy column (Good practice to keep dataframe clean)
fifa = fifa.drop(columns=['Team & Contract'])

# Check the result
print(fifa[['Team', 'Contract Duration']].head())
```

```python
# Only run this if you want to split the years further
# Note: This might fail on rows that say "On Loan" or "Free", so we use errors='ignore' logic implicitly by checking patterns
contract_split = fifa['Contract Duration'].str.split(' ~ ', expand=True)

# Assign safely
fifa['Contract Start'] = contract_split[0]
fifa['Contract End'] = contract_split[1]

print(fifa[['Team', 'Contract Start', 'Contract End']].head())
```
